```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
def even_odd(x):
    if x % 2 == 0:
        return 'Even Number'
    else:
        return 'Odd Number'
```

```python
df['Age_even_odd'] = df['Age'].apply(even_odd)
```

```python
df.head()
```

```python
def gender_number(x):
    if x == 'male':
        return 1
    else:
        return 0
```

```python
df['Gender_number'] = df['Sex'].apply(gender_number)
```

```python
df.head()
```

```python
df['half_fare'] = df['Fare'].apply(lambda x:x/2)
```

```python
df.head()
```
